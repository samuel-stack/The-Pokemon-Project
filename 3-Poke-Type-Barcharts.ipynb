```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

% matplotlib inline
```

```python
poke_pivot = pd.read_csv('./Data/poke_pivot.csv', index_col = 0)
advantages = pd.read_csv('./Data/advantages.csv', index_col = 0)
```

** Getting those official Pokemon type colors**

```python
poke_palette = { "Bug": "#A8B820" , "Dark"    : "#705848" , "Dragon" : "#7038F8" , "Electric": "#F8D030" ,
              "Fairy" : "#EE99AC" , "Fighting": "#C03028" , "Fire"   : "#F08030" , "Flying"  : "#A890F0" ,
              "Ghost" : "#705898" , "Grass"   : "#78C850" , "Ground" : "#E0C068" , "Ice"     : "#98D8D8" ,
              "Normal": "#A8A878" , "Poison"  : "#A040A0" , "Psychic": "#F85888" , "Rock"    : "#B8A038" ,
              "Steel" : "#B8B8D0" , "Water"   : "#6890F0" ,}
```

```python
sns.palplot(sns.color_palette(poke_palette.values()));
```

```python
# Will keep using this same plot to look at individual types
def type_plot(poke_type, poke_pivot = poke_pivot, palette = poke_palette):
    plt.figure(figsize=(15,10))
    plt.xticks(rotation='45',fontsize = 12)
    O_type = pd.DataFrame(poke_pivot.loc[poke_type]).sort_values(poke_type, ascending = False)
    O_type.drop('Total', inplace = True)
    ax = sns.barplot(x=O_type.index  ,y = O_type[poke_type], palette = palette)
    ax.set_ylabel('Occurances', fontsize = 15)
    plt.show();
    print('Total : ', poke_pivot[poke_type].sum())
```

What is a normal Pokémon anyways?  Is it just like a Pokémon that is so average it has not special talents?  We can see that there are over 150 Pokemon with a typing of Normal, and close to ~80% of them are pure normal types and another ~15% of the remaining are those Normal/Flying pairings so it just leaves about ~5% currently of pairings that are normal and some other combination.


Normal Pokémon are in themselves neutral, Normal type Attacks are only super effective against fighting types  and consistent against all other types, but there are no attacking types that are super-effective against normal types.  While this would lead one to believe there could be a defensive advantage with type pairing with normal the way damage is calculated there is not.

Take a fire Pokémon, a pure fire type has a times 2 weakness from water attacks.  Fire/Normal Pokémon would still have a times 2 weakness from water attacks.

What I'm saying is theres not a defensive advantage of pairing any type with normal and for the majority of the times, normal type moves can be learned by all Pokémon so theres no real offensive advantage.   

I guess what I'm saying is theres no reason to investigate these normal pairing combos.
